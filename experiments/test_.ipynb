## Autogluon Tests 

```python
import os

path_inference = '/content/drive/MyDrive/results/inference'
path_columns_types = '/content/drive/MyDrive/results/no_inference'


os.makedirs(path_inference)
os.makedirs(path_columns_types)
```

```python
from autogluon.text import TextPredictor
from autogluon.multimodal import MultiModalPredictor
import pandas as pd 
import json
import os

#Fraction representing the size of the validation set
holdout_frac = 0.2

#Optimization Metrics
metric_bina ='f1'
metric_multi = 'acc'
metric_regr='root_mean_squared_error'

#Times Test
time_limit = [5*60,15*60]

path_inference = '/content/drive/MyDrive/results/inference'
path_columns_types = '/content/drive/MyDrive/results/no_inference'
```

#### Auxiliary Functions

```python
def transform_types(columns_type):
  result = {}
  for name,value in columns_type.items():
    if value == 'int' or value =='float':
      result [name] = 'numerical'
    elif value == 'category':
      result [name] = 'categorical'
    else:
      result [name] = 'text'
  return result  

def preprocess(df):
    df = df.copy(deep=True)
    df.loc[:, 'Reviews'] = pd.to_numeric(df['Reviews'].apply(lambda ele: ele[:-len(' out of 5 stars')]))
    df.loc[:, 'Ratings'] = pd.to_numeric(df['Ratings'].apply(lambda ele: ele.replace(',', '')[:-len(' customer reviews')]))
    return df 


      
```

#### Run Function 

```python

def run(dataset_names, autogluon_predictor ,optimization_metric, time_limit, path_results = '/content/drive/MyDrive/results', enable_type_inference = True):
  dict_ = {}
  y_s = {}
  for name in dataset_names:
    print(name)
    try:
      dataset = HAutoMLBench.get_dataset(name)
      train_data,test_data = dataset.loader_func(dataset)
      info = dataset.info
      if name =='google-guest': 
        target = info['targets'][0]
      else:
        target = info['targets']
      task = info['task']
      column_types = transform_types(info['columns_types'])

      if name =='women-clothing' or name =='pub-health':
        if name =='women-clothing':
          train_data = train_data.loc[train_data[target] != 'Chemises', :].reset_index(drop =True)
        train_data = train_data[train_data[target].notna()]
        test_data = test_data[test_data[target].notna()]
      
      # print(train_data.isnull().sum())
      if name == 'spanish-wine' and not enable_type_inference:
        train_data = train_data[train_data['acidity'].notna()]
        train_data = train_data[train_data['body'].notna()]
        test_data = test_data[test_data['acidity'].notna()]
        test_data = test_data[test_data['body'].notna()]
      
      if name == 'price-book' and not enable_type_inference:
        train_data = preprocess(train_data)
        test_data = preprocess(test_data)
      
      test_data_nolab = test_data.drop([target],axis= 1) 
      print("Fit")
      if enable_type_inference:
        if name =='google-guest': 
          predictor = autogluon_predictor(label= target, eval_metric = optimization_metric, problem_type = task).fit(train_data, time_limit=time_limit, holdout_frac= holdout_frac)
        else:
          predictor = autogluon_predictor(label= target, eval_metric = optimization_metric).fit(train_data, time_limit = time_limit, holdout_frac = holdout_frac)
      else:
        print(column_types)
        if name =='google-guest': 
          predictor = autogluon_predictor(label= target, eval_metric = optimization_metric, problem_type = task).fit(train_data, column_types = column_types, time_limit=time_limit, holdout_frac= holdout_frac)
        else: 
          predictor = autogluon_predictor(label= target, eval_metric = optimization_metric).fit(train_data, column_types = column_types ,time_limit=time_limit,holdout_frac= holdout_frac)

      print("Predict")
      predictions = predictor.predict(test_data_nolab)
      
      
      list_y_true = test_data[target].tolist()
      list_y_pre = predictions.tolist()
    
      print('Benchmark Evaluate')
      
      HAutoMLBench.evaluate(name, list_y_true ,list_y_pre,save_path= path_results, name_archive = f'result_{time_limit}' )
      
      path_y = os.path.join(path_results,f'y_gpu{time_limit}.json')

      try:
        with open(path_y,'r') as fp:
          y_s = json.load(fp)
          y_s[name] = list_y_pre
      except:
          y_s[name] = list_y_pre

      with open(path_y,'w') as fp:
        json.dump(y_s,fp)
    except Exception as error:
      print(f'Error with {name}')   
      print(error) 
```

#### Model Definitions

```python
def binary(time_limit, optimization_metric ,path_results, enable_type_inference):
  bin = HAutoMLBench.filter(task ='binary')
  print("Datasets Binary class")
  print(bin)
  print("Init Binary Classification test")
  
  run(bin[0:3], TextPredictor, optimization_metric, time_limit, path_results, enable_type_inference)
  run(bin[3:6], MultiModalPredictor, optimization_metric, time_limit, path_results, enable_type_inference)
  run(bin[6:8], TextPredictor, optimization_metric, time_limit, path_results, enable_type_inference)
  run(bin[8:9], MultiModalPredictor, optimization_metric, time_limit, path_results, enable_type_inference)

def multiclass(time_limit, optimization_metric, path_results, enable_type_inference):
  multi = HAutoMLBench.filter(task ='multiclass')
  print("Datasets Multiclass")
  print(multi)
  print("Init Multiclass Classification test")
  run(multi[0:2], TextPredictor, optimization_metric , time_limit, path_results, enable_type_inference)
  run(multi[2:5], MultiModalPredictor, optimization_metric, time_limit, path_results, enable_type_inference)
  run(multi[5:6], TextPredictor, optimization_metric , time_limit, path_results,  enable_type_inference)
  run(multi[6:7], TextPredictor, optimization_metric , time_limit, path_results,  enable_type_inference)
  run(multi[7:9], TextPredictor, optimization_metric , time_limit, path_results,  enable_type_inference)



def regression(time_limit, optimization_metric, path_results, enable_type_inference):
  regress = HAutoMLBench.filter(task ='regression')
  print("Datasets Regression")
  print(regress)
  print("Init Regression test")
  run(regress[0:2], MultiModalPredictor, optimization_metric, time_limit, path_results, enable_type_inference)
  run(regress[2:4], TextPredictor, optimization_metric, time_limit, path_results, enable_type_inference)
  run(regress[4:5], MultiModalPredictor, optimization_metric, time_limit, path_results, enable_type_inference)

```
